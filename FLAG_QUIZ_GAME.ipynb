```python
import pygame
import os
import random
import sys

pygame.init()
WIDTH,HEIGHT=800,600
WHITE=(255,255,255)
BLACK=(0,0,0)
LIGHT_ORANGE = (255, 229, 180)
FONT = pygame.font.SysFont('arial',28)
BIG_FONT=pygame.font.SysFont('arial',48)
clock = pygame.time.Clock()

FLAG_FOLDER = r"C:/Users/ranja/OneDrive/Pictures/flags"

screen = pygame.display.set_mode((WIDTH,HEIGHT))
pygame.display.set_caption("üåçFLAG QUIZ GAME")

def draw_text(text,font,color,surface,x,y):
    label = font.render(text, True, color)
    surface.blit(label,(x,y))

def wait_for_key():
    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit();sys.exit()
            if event.type == pygame.KEYDOWN:
                return

def show_welcome():
    screen.fill(LIGHT_ORANGE)
    draw_text("üåç Welcome to the Flag Quiz Challenge!",BIG_FONT,BLACK,screen, 60,100)
    draw_text("press any key to start...",FONT,BLACK,screen, 280,300)
    pygame.display.flip()
    wait_for_key()

def load_questions():
    questions = [
        {
            'image':os.path.join(FLAG_FOLDER, 'australia.png'),
            'answer': 'AUSTRALIA',
            'options': ['UK','USA','AUSTRALIA','NEW ZEALAND']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'belgium.png'),
            'answer': 'BELGIUM',
            'options': ['GERMANY','BELGIUM','SPAIN','TANZANIA']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'brazil.png'),
            'answer': 'BRAZIL',
            'options': ['BRAZIL','SPAIN','ITALY','AFHGANISTAN']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'ireland.png'),
            'answer': 'IRELAND',
            'options': ['MONGOLIA','UGANDA','CHILE','IRELAND']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'italy.png'),
            'answer': 'ITALY',
            'options': ['KAZAKHSTAN','POLAND','ITALY','MOROCCO']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'kenya.png'),
            'answer': 'KENYA',
            'options': ['GHANA','KENYA','NIGERIA','SOMALIA']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'singapore.png'),
            'answer': 'SINGAPORE',
            'options': ['THAILAND','SINGAPORE','INDONESIA','TAIWAN']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'sweden.png'),
            'answer': 'SWEDEN',
            'options': ['NORWAY','POLAND','SWEDEN','FRANCE']
        },
         {
            'image':os.path.join(FLAG_FOLDER, 'switzerland.png'),
            'answer': 'SWITZERLAND',
            'options': ['SWITZERLAND','POLAND','LUXUMBERG','MEXICO']
        }

    ]
    import random
    random.shuffle(questions)
    return questions

def show_flag_question(q,score):
    flag_img = pygame.image.load(q['image'])
    flag_img = pygame.transform.scale(flag_img,(300,200))
    selected = None
    answered = False

    while not answered:
        screen.fill(LIGHT_ORANGE)
        screen.blit(flag_img,((WIDTH - 300)// 2,50))
        draw_text(f"score: {score}",FONT,BLACK,screen,20,20)

        for i , option in enumerate(q['options']):
            draw_text(f"{i+1}.{option}", FONT,BLACK,screen,200,300+i*50)

        pygame.display.flip()
        clock.tick(60)
        
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key in [pygame.K_1, pygame.K_KP1]:
                    selected = q['options'][0]
                elif event.key in [pygame.K_2, pygame.K_KP2]:
                    selected = q['options'][1]
                elif event.key in [pygame.K_3, pygame.K_KP3]:
                    selected = q['options'][2]
                elif event.key in [pygame.K_4, pygame.K_KP4]:
                    selected = q['options'][3]

                if selected:
                    return selected == q['answer']

def show_result(is_correct,correct_answer):
    screen.fill(LIGHT_ORANGE)
    if is_correct:
        draw_text("üéâ Correct!", BIG_FONT, (0, 150, 0), screen, 300, 200)
    else:
        draw_text("‚ùå Wrong!", BIG_FONT, (200, 0, 0), screen, 320, 200)
        draw_text(f"Correct answer: {correct_answer}", FONT, BLACK, screen, 280, 270)
    draw_text("Press any key to continue...", FONT, BLACK, screen, 250, 400)
    pygame.display.flip()
    wait_for_key()

def show_game_over(score, total):
    screen.fill(LIGHT_ORANGE)
    draw_text("üèÅ Game Over!", BIG_FONT, BLACK, screen, 300, 150)
    draw_text(f"Your Score: {score} / {total}", FONT, BLACK, screen, 320, 250)
    draw_text("Press any key to play again or ESC to exit", FONT, BLACK, screen, 170, 350)
    pygame.display.flip()

    while True:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit(); sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_ESCAPE:
                    pygame.quit(); sys.exit()
                else:
                    return

while True:
    show_welcome()
    questions = load_questions()
    score = 0

    for q in questions:
        correct = show_flag_question(q,score)
        if correct:
            score += 1
        show_result(correct, q['answer'])

    show_game_over(score,len(questions))

    
        
    
            

```
